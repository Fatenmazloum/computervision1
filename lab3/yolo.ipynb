Bounding Boxes: The coordinates of the bounding boxes around the detected objects. These are represented as a tensor with shape [num_detections, 4], where each row corresponds to a bounding box defined by its [x_min, y_min, x_max, y_max] coordinates.

Confidence Scores: Each bounding box also has an associated confidence score that tells you how confident the model is in the detection. The confidence score is typically included in the results.boxes as well.
 It is a probability value (a number between 0 and 1), where: A higher confidence score (close to 1) indicates that the model is highly confident in its prediction.A lower confidence score (closer to 0) means the model is less sure about the prediction.

Class IDs: Each detection corresponds to a class ID that indicates which object category was detected (like "person," "car," etc.). This is typically stored in results.boxes as part of the result.

(xmin, ymin) are the coordinates of the top-left corner of the bounding box.

(xmax, ymax) are the coordinates of the bottom-right corner of the bounding box.

```python
#we will draw (plot) boxes
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches #patch is shape(circle,recta...)
def plotboxes(path,results):
    image=Image.open(path)
    plt.figure(figsize=(12,8))
    plt.imshow(image)
    ax=plt.gca()#get current axes of the image
    #plot each boudning box
    for box in results[0].boxes:# In each iteration, box is a single bounding box object
        xyxy=box.xyxy[0]#tensor of coordinates[numder of objects dedicated,4]
        x1,y1,x2,y2=xyxy.numpy()
            #When you need to extract values from a PyTorch tensor and use them in typical Python operations (e.g., mathematical operations, visualization), you may need to convert the tensor to a NumPy array.
        rec=patches.Rectangle((x1,y1),x2-x1,y2-y1,linewidth=1,edgecolor='r',facecolor='none')
            #matplotlib.patches.Rectangle(xy, width, height, angle=0.0, **kwargs)
            #facecolor:fill
        ax.add_patch(rec)#patches need to be added to a specific axes in the figure.not plt
        cls=int(box.cls[0])
        conf=box.conf[0]
        label=f'{results[0].names[cls]} {conf:.2f}'
            #f for formatting f'hi{variable}'
        plt.text(x1,y1,label,color='white',fontsize=10)
            #plt.text() function is used to place text annotations at specific coordinates in a plot.

    plt.axis('off')
    plt.tight_layout()
    plt.show()




plotboxes(path,results)



 
```

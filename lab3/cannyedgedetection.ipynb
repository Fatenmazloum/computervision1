```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img=cv.imread(r"C:\Users\asus\Desktop\Computer Vision\lab3\sammy_face.jpg")
image=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(image,cmap='gray')
```

```python
edges=cv.Canny(image=image,threshold1=127,threshold2=127)
edges
#each array is edge of the picture
```

```python
plt.imshow(edges)
```

```python
blur=cv.blur(image,ksize=(5,5))
edges=cv.Canny(image=blur,threshold1=127,threshold2=127)
plt.imshow(edges)
```

```python
#how to choose best threshold to dive me best edges detection
med=np.median(image)
lower=int(max(0,0.7*med))
upper=int(min(255,1.3*med))
```

```python
edges=cv.Canny(image=image,threshold1=upper,threshold2=lower)
plt.imshow(edges)
```

```python
#blur to decrease noise
blur=cv.blur(image,ksize=(5,5))
edges=cv.Canny(image=blur,threshold1=upper,threshold2=lower)
plt.imshow(edges)
```

for corner detection gray and float

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
imag=cv.imread(r"C:\Users\asus\Desktop\Computer Vision\lab3\book.png")
image=cv.cvtColor(imag,cv.COLOR_BGR2RGB)
img=cv.cvtColor(image,cv.COLOR_RGB2GRAY)
plt.imshow(img,cmap='gray')
img=np.float32(img)#for harris should be float and one channel
corner=cv.cornerHarris(src=img,blockSize=7,ksize=3,k=0.04)
corner=cv.dilate(corner,None)
image[corner>0.01*corner.max()]=[255,0,0]#or imag ,in plt BGR make problem in color
plt.imshow(image)#or iamge
```

```python

pict=cv.imread(r"C:\Users\asus\Desktop\Computer Vision\lab3\book.png")
picg=cv.cvtColor(pict,cv.COLOR_BGR2GRAY)
plt.imshow(picg,cmap='gray')
contour,her=cv.findContours(picg,cv.RETR_CCOMP,cv.CHAIN_APPROX_SIMPLE)
len(contour)
her

```

```python
external=np.zeros(picg.shape)
for i in range(len(contour)):
    if her[0][i][3]==-1:
        cv.drawContours(external,contour,i,255,-1)
plt.imshow(external,cmap='gray')
```

```python
internal=np.zeros(picg.shape)
for i in range(len(contour)):
    if her[0][i][3]!=-1:
        cv.drawContours(internal,contour,i,255)
plt.imshow(internal,cmap='gray')
```

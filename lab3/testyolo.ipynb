```python
from ultralytics import YOLO
model=YOLO('yolov5s.pt')
path=r"C:\Users\asus\Desktop\Computer Vision\lab3\yolo_test_image.jpg"
results=model(path)#results is a container we have only one picture so result[0] predictions for first image each row in the tensor flow is a rectangle of the detected object
results[0]
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
def plot(path,results):
    image=Image.open(path)
    plt.figure(figsize=(18,12))
    plt.imshow(image)
    ax=plt.gca()
    for box in results[0].boxes:#each box have coord as tensor flow[number of objects dedicted,4],conf,cls
        xyxy=box.xyxy[0]
        x1,y1,x2,y2=xyxy.numpy()
        rec=patches.Rectangle((x1,y1),x2-x1,y2-y1,edgecolor='r',linewidth=1,facecolor='none')
        ax.add_patch(rec)
        cls=int(box.cls[0])
        conf=box.conf[0]
        name=f'{results[0].names[cls]}{conf:.2f}'
        plt.text(x1,y1,name,color='white',fontsize=10)
   # plt.axis('off')
   #plt.tight_layout()
    plt.show()
    
plot(path,results)



        
```

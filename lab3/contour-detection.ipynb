Edge: A sharp change in intensity (shows the boundary of an object).

Corner: A point where two edges meet (sharp angle).

Contour: A continuous boundary or outline of an object formed by edges.

So, edges and corners help identify parts of objects, while contours give us the full shape of the object.

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image=cv.imread(r"c:\Users\asus\Downloads\TheAIEngineers-Notebooks-master\ComputerVisionNotebooks\DATA\internal_external.png")
gray=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
contours,heir=cv.findContours(gray,cv.RETR_CCOMP,cv.CHAIN_APPROX_SIMPLE)
print(len(contours))
heir
```

```python
gray.shape
```

```python
#empty black image same shape as image
externalcontour=np.zeros(gray.shape)
for i in range (len(contours)):
    if heir[0][i][3]==-1:#means no contour inside image so external contour
        cv.drawContours(externalcontour,contours,i,255)
plt.imshow(externalcontour,cmap='gray')


```

```python
#empty black image same shape as image
externalcontour=np.zeros(gray.shape)
for i in range (len(contours)):
    if heir[0][i][3]==-1:#means no contour inside image so external contour
        cv.drawContours(externalcontour,contours,i,255,-1)
plt.imshow(externalcontour,cmap='gray')
```

```python
#empty black image same shape as image
internalcontour=np.zeros(gray.shape)
for i in range (len(contours)):
    if heir[0][i][3]!=-1:#means no contour inside image so external contour
        cv.drawContours(internalcontour,contours,i,255)
plt.imshow(internalcontour,cmap='gray')
```

```python
test=cv.imread(r"c:\Users\asus\Downloads\TheAIEngineers-Notebooks-master\ComputerVisionNotebooks\DATA\internal_external.png")
edge=cv.cvtColor(test,cv.COLOR_BGR2GRAY)
plt.imshow(edge,cmap='gray')
cann=cv.Canny(image=edge,threshold1=300,threshold2=300)
plt.imshow(cann,cmap='gray')
```

```python
corner=cv.imread(r"c:\Users\asus\Downloads\TheAIEngineers-Notebooks-master\ComputerVisionNotebooks\DATA\internal_external.png")
gr=cv.cvtColor(corner,cv.COLOR_BGR2GRAY)
plt.imshow(gr,cmap='gray')
floatgr=np.float32(gr)
final=cv.cornerHarris(src=floatgr,blockSize=8,ksize=3,k=0.04)
final=cv.dilate(final,None)
corner[final>0.01*final.max()]=[255,0,0]
plt.imshow(corner)
```

```python
rgbimage=cv.imread(r"C:\Users\asus\Downloads\TheAIEngineers-Notebooks-master\ComputerVisionNotebooks\DATA\rgb_image.jpg")
print(rgbimage.shape)
plt.imshow(rgbimage)
```

```python
rgbimage=cv.cvtColor(rgbimage,cv.COLOR_BGR2RGB)#convert to RGB so BGR2RGB 
print(rgbimage.shape)
plt.imshow(rgbimage)
print(rgbimage)#3 matrix  each matrix  integers between 0 and 255 first intensity of red
```

```python
#image is made up of arrays now split red array alone and green and blue
(R,G,B)=cv.split(rgbimage)
R.shape
```

```python
cv.namedWindow('Blue',cv.WINDOW_NORMAL)#create a window for blue color named blue
cv.namedWindow('Red',cv.WINDOW_NORMAL)
cv.namedWindow('Green',cv.WINDOW_NORMAL)
```

```python
#now display pictures after creating window for each 
cv.imshow("Blue",B)#name of window Blue for B array (1D array)
cv.imshow("Red",R)
cv.imshow("Green",G)
```

```python
#save 
cv.imwrite("red.jpg",R)
cv.imwrite("green.jpg",G)
cv.imwrite("blue.jpg",B)
```

Drawing


```python
import numpy as np
#canvas is an empty page 
canvas=np.zeros((300,300,3),dtype="uint8")# picture 3 dimensions RGB with height and width
#canvas is picture here
# empty black image
```

```python
green=(0,255,0)#green is on middle so 255 is on middle 
cv.line(canvas,(0,0),(300,300),green)#start at 0,0 and ends at 300 300 so like diagonal
cv.namedWindow('canva',cv.WINDOW_NORMAL)
cv.imshow("Canvas",canvas)#to see image (display)
cv.waitKey(0)# close picture by pressing any key
cv.destroyAllWindows()
cv.imwrite('canva.jpg',canvas)
#(0,0) top left where first pixel is located at(0,0)
#(300,0) top right
#(0,300)bottom left
#(300,300)bottom right
```

```python
#colors that canvas take is BGR
red2=(0,0,255)
cv.line(canvas,(299,0),(0,299),red2)
cv.imshow('canva2',canvas)
cv.waitKey(0)
```

Draw Rectangle

```python
cv.rectangle(canvas,(20,200),(60,150),red2,5)#5 is thickness iam plottng rectangle inside canvas image at index
cv.imshow('ractangle',canvas)
cv.waitKey(0)
```

```python
blue=(255,0,0)
cv.rectangle(canvas,(30,70),(77,90),blue,-10)#- thickness: means filled color , image name then starting point and ending point then color and thickness
cv.imshow('final2',canvas)
cv.waitKey(0)
```

```python
#now drawig circle  
(centerx,centery)=(canvas.shape[1]//2 , canvas.shape[0]//2) #width/2 and height/2
white=(255,255,255)#black all zero and white all 255 ,red and blue and green depends
print(centery,centerx)
for r in range(0,175,25):
    cv.circle(canvas,(centerx,centery),r,white)#starting point, center and radius
cv.imshow('circle',canvas)
cv.imwrite("circle.jpg",canvas)
cv.waitKey(0)
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image1=cv.imread(r"C:\Users\asus\Downloads\TheAIEngineers-Notebooks-master\ComputerVisionNotebooks\DATA\image1.png")
cv.imshow("BGRimage",image1)
cv.waitKey(0)
image1.shape
```

we will make filter(gradiant) to it to find edges of picture.

```python
#to make edge detection we should convert picture to gray scale from BGR
image=cv.cvtColor(image1,cv.COLOR_BGR2GRAY)
cv.imshow("grayimage",image)
cv.waitKey(0)
image.shape
```

```python
#create prewitt filter which is  array with 3*3
#prewitt filter for edge detection
kernelx=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
kernely=np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
#apply prewitt filter to image
filterimagex=cv.filter2D(image,-1,kernelx)
filterimagey=cv.filter2D(image,-1,kernely)
#Gradiant magnitude
filterimage=np.hypot(filterimagey,filterimagex) #hypot is square root of sum of both gradiants x and y so radical gradiant y square + gradiant x square
#filterimage=filterimage.astype("int")#from float to int beacuse intensity as integer range between 0 and 255
plt.imshow(filterimage,cmap="gray")
```

The Sobel operator is a simple edge detection technique that calculates the gradient of image intensity at each pixel.
Result: The Sobel operator provides a gradient map highlighting edges where the intensity changes rapidly

```python
#sobel filter which is 3*3 matrix
kernelx=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
kernely=np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
filterimagex=cv.filter2D(image,-1,kernelx)
filterimagey=cv.filter2D(image,-1,kernely)
filterimage=np.hypot(filterimagey,filterimagex) #hypot is square root of sum of both gradiants x and y so radical gradiant y square + gradiant x square
#filterimage=filterimage.astype("int")
plt.imshow(filterimage,cmap="gray")
```
